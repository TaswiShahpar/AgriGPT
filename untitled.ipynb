```python
import pandas as pd

# Load the dataset
df = pd.read_csv('questionsv4.csv')

# Display the first few rows
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# Vectorize the questions
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['questions'])

# Fit the Nearest Neighbors model
nn = NearestNeighbors(n_neighbors=1, algorithm='brute', metric='cosine')
nn.fit(X)
```

```python
import gradio as gr

def get_answer(query):
    query_vec = vectorizer.transform([query])
    distance, index = nn.kneighbors(query_vec)
    return df['answers'].iloc[index[0][0]]

# Create the Gradio interface
iface = gr.Interface(fn=get_answer,
                     inputs="text",
                     outputs="text",
                     title="Farming Chatbot",
                     description="Ask any farming-related question!")

# Launch the interface
iface.launch()
```
